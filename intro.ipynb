```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resizing your graph
plt.figure(figsize=(5,3), dpi=300)

#Plotting the points and giving it a specific label/name
#Keyword argument notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, linestyle='--',marker='.',markersize=10, markeredgecolor='blue')

#Shorthand notation
plt.plot(x,y, 'b*--', label='2x')

#Adding a title and giving it specific features
plt.title('Our First Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})

#Labelling the axis
plt.xlabel('X axis')
plt.ylabel('Y axis')

#Line two
#Choosing the intervals we want the graph to be plotted at
x2 = np.arange(0,4.5,0.5)

#Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

#Plot part of the graph as dashline
plt.plot(x2[4:], x2[4:]**2, 'r--')

#Configuring the ticks on both axis
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

#Adding a legend to the graph
plt.legend()

#Saving the graph
plt.savefig('myfirstgraph.png', dpi=400)

plt.show()
```

###Bar charts

```python
labels =['A','B', 'C']
values =[1,4,2]

bars = plt.bar(labels,values)

#Setting the hatch for the specific bars
#Short hand method
patterns=['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
#Long route
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```
