```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(5)  #Assuming there are five students
y = (20, 35, 30, 35, 27)  #Their test scores

#Title of graph
plt.title('Graph of Student\'s marks')
plt.bar(x,y)

#x and y labels
plt.xlabel('Number of students')
plt.ylabel('Marks/Score')

plt.legend()
plt.show()

#Plotting the scatter graph
plt.scatter(x,y)

#x and y labels
plt.xlabel('Number of students')
plt.ylabel('Marks/Score')

plt.show()
```
