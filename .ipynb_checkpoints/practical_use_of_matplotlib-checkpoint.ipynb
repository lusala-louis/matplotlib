```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')

#Sizing the graph
plt.figure(figsize=(8,5), dpi = 300)

#Ploting the graphs of two countries from the csv data file
plt.plot(gas.Year, gas.USA,'b.-', label='United States')
plt.plot(gas.Year, gas.Canada,'y.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')

#A different method to do it
#countries = ['Australia', 'USA', 'Canada', 'South Korea']
#for country in gas:
#    if country in countries:
#        plt.plot(gas.Year, gas[country], marker='.')

#Adding a title
plt.title('Gas Prices Over Time', fontdict={'fontweight':'bold','fontsize':18})

#Giving the x and y axis labels
plt.xlabel('Year', fontdict={'fontweight':'bold','fontsize':12})
plt.ylabel('Prices per gallon(USD)', fontdict={'fontweight':'bold','fontsize':12})

#Setting the interval of the years in the x axis to two
plt.xticks(gas.Year[::2])

#A different way to plot the legend of various countries
#for country in gas:
#   if country in countries:
#        plt.legend(gas[country])

plt.legend()

plt.savefig('Gas_price_figure.png',dpi=300)
        
plt.show()
```

```python
#Sizing the graph
plt.figure(figsize=(8,5), dpi = 300)

for country in gas.columns[1:]:
    plt.plot(gas.Year, gas[country], marker='.')
    
#Adding a title
plt.title('Gas Prices Over Time')

#Giving the x and y axis labels
plt.xlabel('Year')
plt.ylabel('Prices per gallon(USD)')

#Setting the interval of the years in the x axis to two
plt.xticks(gas.Year[::2])

for country in gas.columns[1:]:
    plt.legend(gas[country])

#plt.legend()
        
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(7)
```

Histogram

```python
bins =[40,50,60,70,80,90,100]

plt.figure(figsize=(6,4), dpi=300)

plt.hist(fifa.Overall, bins = bins)

plt.xticks(bins)

plt.ylabel('Number of players', fontdict={'fontweight':'bold', 'fontsize':12})
plt.xlabel('Skill level',fontdict={'fontweight':'bold', 'fontsize':12})

plt.title('Distribution of Player Skills in FIFA 2018', fontdict={'fontweight':'bold', 'fontsize':14})
plt.savefig('Player_skills_graph.png',dpi=300)

plt.show()
```

Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(4,3), dpi=300)
labels =['Left','Right']
plt.pie([left, right], labels = labels, autopct ='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
#Remove the lbs string from the weights column and convert the weights to integers
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

#plt.figure(figsize=(4,3), dpi=300)
plt.style.use('ggplot')

light = fifa[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight > 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125 - 150', '150 - 175', '175 - 200', 'Over 200']
explode = (0.4,0.2,0,0,0.4)

plt.pie(weights, labels = labels, autopct='%.2f %%', pctdistance=0.8, explode = explode)
plt.title('Weight Distibution of FIFA Players (in lbs)')

plt.show()
```

```python
plt.figure(figsize=(5,8))
plt.style.use('default')

barcelona = fifa.loc[fifa.Club =='FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels =['FC Barcelona', 'Real Madrid', 'New\n England\n Revolution']

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Ratings')

boxes = plt.boxplot([barcelona, madrid, revs], labels =labels, medianprops={'linewidth':2}, patch_artist=True)
for box in boxes['boxes']:
    #Set the edge color
    box.set(color='#4286f4', linewidth=2)
    #set the fill
    box.set(facecolor='#e0e0e0')

plt.savefig('Football_team_comparison.png',dpi=300)
plt.show()
```
